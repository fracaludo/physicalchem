```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# === File path ===
file_path = r"C:\Users\emmak\github\TP-GC-BA6\Physical\Challenge 2\UVvisgraph.xlsx"
folder_path = os.path.dirname(file_path)

# === Read the Excel file ===
df = pd.read_excel(file_path, decimal=",")
df.rename(columns={df.columns[0]: "Wavelength"}, inplace=True)

# === Ensure column names are strings ===
df.columns = df.columns.map(str)

# === Extract only normalized sample columns ===
norm_cols = [col for col in df.columns if "norm" in col]

# === Plotting ===
# Using the 'Blues' colormap for shades of blue
colors = plt.cm.Blues(np.linspace(0.2, 0.9, len(norm_cols)))

plt.figure(figsize=(10, 6))
for col, color in zip(norm_cols, colors):
    plt.plot(df["Wavelength"], df[col], label=col.replace("norm", "").strip(), color=color)

# === Customize plot ===
plt.xlabel("Wavelength [nm]")
plt.ylabel("Normalized Absorbance [a.u.]")
#plt.title("Normalized UV-Vis Absorbance vs. Wavelength")
plt.legend(title="Sample ", loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.gca().set_facecolor('#f9f9f9')
plt.minorticks_on()
plt.tick_params(which='both', direction='in', top=False, right=False)
plt.tick_params(which='minor', length=4, color='gray')
plt.tick_params(which='major', length=6)

# === Save the plot ===
def save_plot(directory, filename="UVvis_Normalized_Plot.png", format="png"):
    os.makedirs(directory, exist_ok=True)
    filepath = os.path.join(directory, filename)
    plt.savefig(filepath, format=format, dpi=600, bbox_inches="tight")
    print(f"âœ… Plot saved to: {filepath}")

save_plot(folder_path)

# === Show plot ===
plt.show()
```
