```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# File path
file_path = "/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/Fluo/1 method.xlsx"
folder_path = '/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/Fluo'

# Read the Excel file
df = pd.read_excel(file_path, sheet_name="normalised ", decimal=",")  # Adjust for European decimal notation

# Rename the first column to 'Wavelength'
df.rename(columns={df.columns[0]: "Wavelength"}, inplace=True)

# Generate a color gradient (shades of green)
colors = plt.cm.Greens(np.linspace(0.2, 1, len(df.columns[1:])))

# Plot all europium concentrations
plt.figure(figsize=(10, 6))
for col, color in zip(df.columns[1:], colors):  # Skip the Wavelength column
    plt.plot(df["Wavelength"], df[col], label=f"{col} µM", color=color)

# Customize the plot
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity [a.u.]")
plt.title("Fluorescence Intensity vs. Wavelength for Different Eu³⁺ Concentrations Method 1 ")
plt.legend(title="Eu³⁺ Concentration [µM]", loc="upper right")
plt.grid(True)

wavelength_to_add = 280  # Specify your desired wavelength here
plt.axvline(x=wavelength_to_add, color='green', linestyle='--', label=f"{wavelength_to_add} nm")

def save_plot(directory, filename="Processed_Plot 1.png", format="png"):
    os.makedirs(directory, exist_ok=True)
    filepath = os.path.join(directory, filename)
    plt.savefig(filepath, format=format, dpi=600, bbox_inches="tight")
    print(f"Plot saved to: {filepath}")

save_plot(folder_path)

# Show the plot
plt.show()

```
