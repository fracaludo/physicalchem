```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import os

# Hill equation definition
def hill_eq(x, n, Ka):
    return (x**n) / (Ka**n + x**n)

# Given data
ycbien = np.array([
    0.074710759,
    0.077817853,
    0.124174434,
    0.083048386,
    0.300341936,
    0.30424532,
    0.984929973,
    1.0
])

xd = np.array([
    2.17e-05,
    3.24e-05,
    4.30e-05,
    5.36e-05,
    6.41e-05,
    7.45e-05,
    9.51e-05,
    1.05e-04
])

# Fit the data to the Hill equation
popt, pcov = curve_fit(hill_eq, xd, ycbien, p0=[1, 1])  # Initial guess for n and Kd
n_opt, Ka_opt = popt  # Optimized parameters

# Generate fitted values
y_fit = hill_eq(xd, *popt)

# Calculate R^2 score
r2 = r2_score(ycbien, y_fit)

# Plotting the original data and fitted curve
plt.figure(figsize=(8, 6))

# Plot: Standard Plot
plt.plot(xd, ycbien, 'bo', color="royalblue", label='Experimental data')
plt.plot(xd, y_fit, 'r-', color="gold", label=f'Hill fit: n={n_opt:.2f}, Ka={Ka_opt:.8e}, RÂ² = {r2:.3f}')
plt.xlabel('[Eu$^{3+}$] in the tube [mol/L]')
plt.ylabel('Normalized Peak intensity [a.u]')
plt.legend()
plt.title('Hill Equation Fit Peak 1')
plt.tight_layout()
plt.gca().set_facecolor('#f9f9f9')

#function to save the plot 
def save_plot(directory, filename="Hill Equation Fit peak1.png", format="png"):
    if not os.path.exists(directory):
        os.makedirs(directory)  # Create directory if it doesn't exist
    filepath = os.path.join(directory, filename)
    plt.savefig(filepath, format=format, dpi=600, bbox_inches="tight")
    print(f"Plot saved to: {filepath}")
save_plot("/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/NMR")

plt.show()

# Print optimized parameters and RÂ²
print(f'Optimized n: {n_opt:.3f}')
print(f'Optimized Ka: {Ka_opt:.8e}')  
print(f'RÂ²: {r2:.3f}')
Kb =1/(Ka_opt)**(1/n_opt)
Kb
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import os

# Hill equation definition
def hill_eq(x, n, Ka):
    return (x**n) / (Ka**n + x**n)

# Original data
ycbien = np.array([
    0.01923769, 0, 0.613234091, 0.625617536, 0.707279568, 0.768591526,
    0.927663995, 0.832193686, 0.953551448, 0.33063962, 
    0.383625061, 0.356617046, 0.704923933, 0.849407819, 1
])

xd = np.array([
    0.00E+00, 1.09E-05, 2.17E-05, 3.24E-05, 4.30E-05, 5.36E-05, 6.41E-05, 
    7.45E-05, 9.51E-05, 1.05E-04, 1.25E-04, 1.65E-04, 
    1.84E-04, 2.03E-04, 2.50E-04
])

# Indices to remove: 3 to 9 => Python indices 2 to 8
indices_to_remove = list(range(2, 9))

# Separate removed and kept data
xd_removed = xd[indices_to_remove]
ycbien_removed = ycbien[indices_to_remove]

xd_fit = np.delete(xd, indices_to_remove)
ycbien_fit = np.delete(ycbien, indices_to_remove)

# Fit all data
popt_all, _ = curve_fit(hill_eq, xd, ycbien, p0=[1, 1], maxfev=5000)
n_all, Ka_all = popt_all
y_fit_all = hill_eq(xd, *popt_all)
r2_all = r2_score(ycbien, y_fit_all)

# Fit partial data
popt_partial, _ = curve_fit(hill_eq, xd_fit, ycbien_fit, p0=[1, 1], maxfev=5000)
n_partial, Ka_partial = popt_partial
y_fit_partial = hill_eq(xd, *popt_partial)
r2_partial = r2_score(ycbien_fit, hill_eq(xd_fit, *popt_partial))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(xd, ycbien, 'o', color="royalblue", label='All data')
plt.plot(xd_removed, ycbien_removed, 'o', color="lightsteelblue", label='Removed data')
plt.plot(xd, y_fit_all, '-', color="palegoldenrod", label=f'Fit (all): n={n_all:.2f}, Ka={Ka_all:.2e}, RÂ²={r2_all:.3f}')
plt.plot(xd, y_fit_partial, '-', color="gold", label=f'Fit (partial): n={n_partial:.2f}, Ka={Ka_partial:.2e}, RÂ²={r2_partial:.3f}')

plt.xlabel('[Eu$^{3+}$] in the tube [mol/L]')
plt.ylabel('Normalized Peak intensity [a.u]')
plt.title('Hill Equation Fit â Peak 7')
plt.legend()
plt.gca().set_facecolor('#f9f9f9')
plt.tight_layout()

# Save plot
def save_plot(directory, filename="Hill Equation Fit peak7.png", format="png"):
    if not os.path.exists(directory):
        os.makedirs(directory)
    filepath = os.path.join(directory, filename)
    plt.savefig(filepath, format=format, dpi=600, bbox_inches="tight")
    print(f"Plot saved to: {filepath}")

save_plot("/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/NMR")

plt.show()

# Print fitted values
print("=== Full Fit ===")
print(f'Optimized n: {n_all:.3f}')
print(f'Optimized Ka: {Ka_all:.8e}')
print(f'RÂ²: {r2_all:.3f}')
print(f'Kb (full): {1 / (Ka_all)**(1/n_all):.3e}')

print("\n=== Partial Fit (values 3â9 removed) ===")
print(f'Optimized n: {n_partial:.3f}')
print(f'Optimized Ka: {Ka_partial:.8e}')
print(f'RÂ²: {r2_partial:.3f}')
print(f'Kb (partial): {1 / (Ka_partial)**(1/n_partial):.3e}')
```
