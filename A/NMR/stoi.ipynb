```python
import pandas as pd
import matplotlib.pyplot as plt
import os 

# Define file path and sheet name
file_path = "/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/NMR/NMR - stoi.xlsx"
sheet_name = "5"  

# Read the Excel file
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Ensure column names are clean
df.columns = df.columns.str.strip()

# Extract the relevant columns
x = df["Eu3+/CaM"]
y = df["CSP OBS/CSP Max"]

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color="royalblue", label="Experimental Data")  
plt.xlabel("[Eu³⁺]/[CaM]")
plt.ylabel(r"$\frac{CSP_{OBS}}{CSP_{MAX}}$")
plt.title(r"$CSP_{OBS}/CSP_{MAX}$ vs [Eu³⁺]/[CaM] Peak 5")
plt.legend(loc="best")  # Automatically positions the legend
plt.tight_layout()
plt.gca().set_facecolor('#f9f9f9')

# Function to save the plot
def save_plot(directory, filename="Stoi Peak 5.png"):
    os.makedirs(directory, exist_ok=True)  # Create directory if it doesn't exist
    filepath = os.path.join(directory, filename)
    plt.savefig(filepath, dpi=600, bbox_inches="tight")
    print(f"Plot saved to: {filepath}")

# Save the plot
save_plot("/Users/ludovica/Documents/Bachelor/BA6/Physical Chemistry/Challenge A/NMR")

# Show the plot
plt.show()
```
